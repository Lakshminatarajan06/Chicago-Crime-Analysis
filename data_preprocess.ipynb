```python
# Split Date and Time
def split_date_time(df, col):

    try:

        df[col]=pd.to_datetime(df[col], format= '%m/%d/%Y %I:%M:%S %p')
        # Extract Date and Time in DataFrame
        date=df[col].dt.date
        time=df[col].dt.time

        # Create new column for Date and Time

        df['Date']=date
        df['Time']=time

        return df
    except Exception as e:
        print(f'The error Occured:{e}')

df=split_date_time(df, 'Date')
    
```

```python
# Change the 'Date' column is of string type
df['Date'] = df['Date'].astype(str)
```

```python
# Split the 'Date' column
split_dates=df['Date'].str.split("-",expand=True)

# Assign the split components to new columns
df['Year']=split_dates[0]
df['Month']=split_dates[1]
df['Day']=split_dates[2]
```
